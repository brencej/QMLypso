```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("E:/Dropbox/github/QMLypso/experiment1_1000shots.csv")
```

```python
plt.figure(figsize=(8,5), dpi=150)
plt.rcParams.update({"font.size":14})
plt.plot(data["par"], data["val"], label="value", lw=2)
plt.plot(data["par"], data["grad"], label="gradient", lw=2)
plt.legend()
plt.title("$<\\Psi|\\sigma_z|\Psi>$")
plt.show()
```

```python
folder = "E:/Dropbox/github/QMLypso/results/"
tk0 = np.loadtxt(folder + "pytketv0.txt")
tk1 = np.loadtxt(folder + "pytketv1.txt")
qs0 = -1*np.loadtxt(folder + "qiskitv0.txt")
qs1 = -1*np.loadtxt(folder + "qiskitv1.txt")
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(tk0)
ax[1].imshow(tk1)
ax[0].set_xlabel("$\Theta_1$")
ax[1].set_xlabel("$\Theta_1$")
ax[0].set_ylabel("$\Theta_2$")
ax[0].set_title("$\\frac{\\partial f}{\\partial \\Theta_1}$")
ax[1].set_title("$\\frac{\\partial f}{\\partial \\Theta_2}$")
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(qs0)
ax[1].imshow(qs1)
plt.show()
```

```python
err = np.sqrt((tk0-qs0)**2 + (tk1-qs1)**2)
```

```python
plt.imshow(err)
plt.xlabel("$\Theta_1$")
plt.ylabel("$\Theta_2$")
plt.colorbar()
plt.title("sqrt(grad(qiskit) - grad(tket))")
plt.show()
```
